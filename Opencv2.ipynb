```python
#import library yang diperlukan dulu ya gaes
import cv2 #ini library opencv nya
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load image atau foto yg mau dipake ya
image = cv2.imread('./image2.jpg') #nah ini path foto yg mau dipake, biasanya setiap komputer pasti beda. jadi kalo error jangan lupa googling ya jan banyak nanya
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#kita ubah ke gray alias abu abu image nya
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
edge = cv2.Canny(gray, 60, 180) #kita cari ujung setiap objek yang ada di image, fungsinya untuk apa? next video wqwq
plt.imshow(edge, cmap="Greys")
```

```python
#saatnya cari kontur image
contours = cv2.findContours(edge, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
#oiya lupa belum di draw wqwqw
#kita dray kontur yg sudah didapatkan supaya enak liatnya
cv2.drawContours(image, contours[0], -1, (0,0,255), thickness =2)
plt.imshow(image)
```
